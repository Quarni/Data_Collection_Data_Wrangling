```r
 df <- USPersonalExpenditure
data <- as.data.frame(df)
data
```

```r
### start solution ####
```

```r
#Compute summary statistics: mean, standard deviation (sd) across time for each of the indicators.

summarise_at(data, vars('1940','1945','1950','1955','1960'), list(~n(), ~mean(.), ~sd(.)))
```

```r
#Create a new column that contains average values for each of the indicators.
colMeans(data)
```

```r
# Another method for culculating Avg or mean columns wise
mutate_at(data, vars('1940','1945','1950','1955','1960'), funs(mean=mean(.)))
```

```r
#Create a new column that contains average values for each of the indicators. row wise mean 
mutate(data, Avg=rowMeans(data))
```

##### This is the most common method of data imputation, where you just replace all the missing values with the mean, median or mode of the column. While this is useful if you’re in a rush because it’s easy and fast, it changes the statistical nature of the data. Not only does this skew our histograms, it also underestimates the variance in our data because we’re making numerous values the exact same (when in reality they evidently would not be).
